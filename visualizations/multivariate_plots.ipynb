```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
cov  = [[1,   0],
        [0, 100]]

multivariate_dist = np.random.multivariate_normal(mean, cov, 10000)
```

```python
g = sns.jointplot(x=multivariate_dist[:,0], y=multivariate_dist[:,1])
```

```python
mean = [0, 0, 0]
cov  = [[1,   0, 0], 
        [0, 100, 0],
        [0,   0, 1]]

multivariate_dist = np.random.multivariate_normal(mean, cov, 10000)

fig = go.Figure(data=[go.Scatter3d(x=multivariate_dist[:,0], y=multivariate_dist[:, 1], z=multivariate_dist[:, 2],
                                   mode='markers')])

# Formatting
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.update_traces(marker_size = 1)
fig.show()
```
